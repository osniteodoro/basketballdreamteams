# **TEAM STATS PER GAME - Estatísticas Time por Jogo**

```python
stats_pergame_1992 = pd.read_csv("../input/usa-dream-teams/1992 - Stats - Team Stats Per Game.csv")
stats_pergame_1996 = pd.read_csv("../input/usa-dream-teams/1996 - Stats - Team Stats Per Game.csv")
stats_pergame_2000 = pd.read_csv("../input/usa-dream-teams/2000 - Stats - Team Stats Per Game.csv")
stats_pergame_2012 = pd.read_csv("../input/usa-dream-teams/2012 - Stats - Team Stats Per Game.csv")
stats_pergame_2016 = pd.read_csv("../input/usa-dream-teams/2016 - Stats - Team Stats Per Game.csv")
```

```python
stats_pergame_1992.query("Club == 'United States'")
                      
```

```python
aux_1992 = stats_pergame_1992
aux_1992['Year'] = '1992'
aux_1996 = stats_pergame_1996
aux_1996['Year'] = '1996'
aux_2000 = stats_pergame_2000
aux_2000['Year'] = '2000'
aux_2012 = stats_pergame_2012
aux_2012['Year'] = '2012'
aux_2016= stats_pergame_2016
aux_2016['Year'] = '2016'


aux_1992.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS"},inplace = True )
aux_1996.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS"},inplace = True )
aux_2000.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS"},inplace = True )
aux_2012.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS"},inplace = True )
aux_2016.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS"},inplace = True )
```

```python

tabela_1 = aux_1992
tabela_2 = aux_1996
tabela_3 = aux_2000
tabela_4 = aux_2012
tabela_5 = aux_2016

resultado = pd.concat([tabela_1, tabela_2, tabela_3, tabela_4, tabela_5])
```

```python
resultado.query("Club == 'United States'")
```

```python
#Seleção com maior número de pontos por jogo e G (total de games)
pontos=resultado[['Club', 'Year','FG','G']][resultado.FG == resultado['FG'].max()]
print(pontos)
```

```python
total_1992 = pd.read_csv("../input/usa-dream-teams/1992 - USA - Totals.csv")
total_1996 = pd.read_csv("../input/usa-dream-teams/1996 - USA - Totals.csv")
total_2000 = pd.read_csv("../input/usa-dream-teams/2000 - USA - Totals.csv")
total_2012 = pd.read_csv("../input/usa-dream-teams/2012 - USA - Totals.csv")
total_2016 = pd.read_csv("../input/usa-dream-teams/2016 - USA - Totals.csv")

total_1992['Year'] = '1992'
total_1996['Year'] = '1996'
total_2000['Year'] = '2000'
total_2012['Year'] = '2012'
total_2016['Year'] = '2016'

total_1992.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS", "2PA": "TWO_PTS_PA", "2P%": "TWO%"},inplace = True )
total_1996.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS", "2PA": "TWO_PTS_PA", "2P%": "TWO%"},inplace = True )
total_2000.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS", "2PA": "TWO_PTS_PA", "2P%": "TWO%"},inplace = True )
total_2012.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS", "2PA": "TWO_PTS_PA", "2P%": "TWO%"},inplace = True )
total_2016.rename(columns={"3PA": "THREE_PTS_PA","3P": "THREE_PTS", "3P%": "THREE%", "2P": "TWO_PTS", "2PA": "TWO_PTS_PA", "2P%": "TWO%"},inplace = True )
```

```python
tabela_1 = total_1992
tabela_2 = total_1996
tabela_3 = total_2000
tabela_4 = total_2012
tabela_5 = total_2016
		
total = pd.concat([tabela_1, tabela_2, tabela_3, tabela_4, tabela_5])
```

```python
#Seleção com 5 maiores pontuadores no total e G (total de games)
pontos=total[['Player', 'Year','PTS', 'G']].nlargest(5,"PTS")
print(pontos)
```
